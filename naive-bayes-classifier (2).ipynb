```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("LA_Crime_Data_2020_2023.csv")
data.head()
```

```python
data['CrmDesc'].unique()
```

```python
len(data['CrmDesc'].unique())
```

```python
def danger_group(x):
    if x=='BURGLARY FROM VEHICLE' or x=='ROBBERY' or x=='BURGLARY' or x=='VEHICLE - STOLEN' or x=='VANDALISM - MISDEAMEANOR ($399 OR UNDER)' or x=='VANDALISM - FELONY ($400 & OVER, ALL CHURCH VANDALISMS)' or x=='SHOPLIFTING - PETTY THEFT ($950 & UNDER)' or x=='THEFT-GRAND ($950.01 & OVER)EXCPT,GUNS,FOWL,LIVESTK,PROD' or x=='THEFT PLAIN - PETTY ($950 & UNDER)' or x=='THEFT FROM MOTOR VEHICLE - PETTY ($950 & UNDER)' or x=='THEFT FROM MOTOR VEHICLE - GRAND ($950.01 AND OVER)' or x=='CRIMINAL THREATS - NO WEAPON DISPLAYED' or x=='TRESPASSING':
        return 'first degree'
    else:
        return 'second degree'

data['danger_crime']=data['CrmDesc'].apply(danger_group)
data['danger_crime']
    
    
```

```python
data['danger_crime'].unique()
```

```python
X = data[['Area','DistrictNo','LAT','LON', 'TimeStamp']]

y = data['danger_crime']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.2, random_state=42)
```

```python
# train a Gaussian Naive Bayes classifier on the training set
# instantiate the model
# initializaing the NB
classifer = GaussianNB()
# training the model
classifer.fit(X_train, y_train)
# testing the model
y_pred = classifer.predict(X_test)
```

```python

y_pred
```

```python
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
print('Training set score: {:.4f}'.format(classifer.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(classifer.score(X_test, y_test)))
```

```python
# Print the Confusion Matrix and slice it into four pieces

cm = confusion_matrix(y_test, y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
print(classification_report(y_test, y_pred))
```
